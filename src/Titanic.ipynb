```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import japanize_matplotlib
%matplotlib inline
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

import numpy as np
```

```python
baseURI = '../data/'
df_testData = pd.read_table(f"{baseURI}/test.tsv")
df_trainData = pd.read_table(f"{baseURI}/train.tsv")
```

```python
class_mapping_sex = {label: idx for idx, label in enumerate(np.unique(df_trainData["sex"]))}
class_mapping_sex
```

```python
df_trainData["embarked"].value_counts()
```

(df_trainData["embarked"])

```python
class_mapping_embarked = {label: idx for idx, label in enumerate(["S","Q","C","nan"])}
class_mapping_embarked
```

```python
df_trainData_copy = df_trainData
```

```python
df_trainData_copy["sex"] = df_trainData_copy["sex"].map(class_mapping_sex)
```

```python
df_trainData_copy
```

```python
df_trainData_copy["embarked"] = df_trainData_copy["embarked"].map(class_mapping_embarked)
```

```python
df_trainData_copy
```

```python
df_trainData = df_trainData_copy
```

```python
df_testData["sex"] = df_testData["sex"].map(class_mapping_sex)
```

```python
df_testData["embarked"] = df_testData["embarked"].map(class_mapping_embarked)
```

```python
y_train = df_trainData[df_trainData.columns[df_trainData.columns == "survived"]]
X_train = df_trainData[df_trainData.columns[df_trainData.columns != "survived"] & df_trainData.columns[df_trainData.columns != "id"]]
```

```python
X_train
```

```python
y_train
```

```python
y_test = df_testData[df_testData.columns[df_testData.columns != "id"]]
```

```python
y_test
```

```python
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier()

lgbm.fit(X_train, y_train)

y_pred = lgbm.predict(y_test)
```

```python
print(y_pred)
```

```python
df_result = pd.DataFrame({ 'id' : df_testData["id"],
                        'y_pred' : y_pred
                         })
```
